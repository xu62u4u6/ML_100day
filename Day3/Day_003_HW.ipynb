### Using Machine Learning to Explore Neural Network Architecture


1.__目標__:<BR>
 如文中所提-__a typical 10-layer network can have $10^{10}$ candidate networks!__，所以目標就是找出最佳的模型配置<BR><BR>
2.__使用技術__<BR>
 是AutoML，是一個有點類似rnn的概念，本輪的回饋會影響下一輪的建議，來找出一個最適當的model架構，目前學習出模型已經可以媲美專家模型<br>
 <br>
 有趣的是算出的模型，含有乘法閥，這在以往的循環模型是不常見的，有趣的是這和最近的發現不謀而合，<br>
 就是乘法閥可能能解決梯度爆炸或是消失的問題，另外就是此模型含有許多通道，可以讓梯度流向這些通道，這也解釋了為何LSTM效果會比RNN好<br><BR>
